```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("vintage.jpg")
imagemine = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

subtracted_image = cv2.subtract(imagemine, 200)

plt.subplot(1, 2, 1)
plt.imshow(imagemine)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(subtracted_image)
plt.title('Subtracted Image (-100)')
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('vintage.jpg')
imagemine = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

multipliedimage = cv2.multiply(imagemine, 2)

plt.subplot(1, 2, 1)
plt.imshow(imagemine)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(multipliedimage)
plt.title('Multiplied Image 2')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('vintage.jpg')
imagemine = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

multipliedimage = cv2.divide(imagemine, 2)
# Display original and divided images
plt.subplot(1, 2, 1)
plt.imshow(imagemine)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(multipliedimage)
plt.title('Divided Image 2')
plt.axis('off')
plt.show()
```
